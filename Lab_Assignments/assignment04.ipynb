## Linear Regression --- In Personal Dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
file_path='../Personal_Datasets/housing.csv'
df=pd.read_csv(file_path)
df
```

```python
plt.scatter(df.avg_area_income,df.Price,color='red',marker='*')
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[['avg_area_income']],df.Price)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
file_paths='../Personal_Datasets/housing1.csv'
d=pd.read_csv(file_paths)
d.head()
```

```python
p=reg.predict(d)
p
d['price']=p
plt.xlabel('avg_area_income',fontsize=20)
plt.ylabel('Price',fontsize=20)
plt.scatter(df.avg_area_income,df.Price,color='red',marker='*')
plt.plot(df.avg_area_income,reg.predict(df[['avg_area_income']]),color='blue')
```
