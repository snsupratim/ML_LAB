## Linear Regression --- In Personal Dataset

```python
import pandas as pd
import seaborn as sns
import numpy as np
from sklearn import preprocessing, svm
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
file_path='../Personal_Datasets/KOLKATA_Housing.csv'
df=pd.read_csv(file_path)
```

```python
df_binary = df[['avg_area_house_age', 'Price']].copy()
df_binary_columns = ['avg_area_house_age', 'Price']
df_binary.head()
```

```python
x = df['avg_area_house_age']
y = df['Price']
```

```python
X = np.array(df_binary['avg_area_house_age']).reshape(-1,1)
Y = np.array(df_binary['Price']).reshape(-1,1)
#df_binary.dropna(inplace=True)
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.33)
```

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
print(lr.score(x_test,y_test))
lr.coef_
lr.intercept_

m=lr.intercept_
b=lr.coef_
print('y=',m,'*X+',b)
```

```python
pred = lr.predict(x_test)
print(pred)
```

```python
plt.scatter(x_train,y_train)
plt.plot(x_test,pred,color='red')
```
