```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("tryout.csv")
# df.head()
df=df[['CRIM','ZN','INDUS','CHAS','NOX','AGE','DIS','RAD','PTRATIO','B','LSTAT','MEDV','TAX']]

df.columns[df.isnull().any()]
```

```python
x=df.drop('TAX',axis=1)
y=df['TAX']
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=1)
```

```python
# let k=69

from sklearn.neighbors import KNeighborsClassifier

model=KNeighborsClassifier(n_neighbors=69)
model.fit(x_train,y_train)
```

```python
y_test_hat=model.predict(x_test)

results=pd.DataFrame({'Actual':y_test,'Predictions':y_test_hat})
results.head(5)
```

```python
from sklearn.metrics import confusion_matrix,recall_score,precision_score

cm=confusion_matrix(y_test,y_test_hat)
print(cm)

from sklearn.metrics import classification_report
print(classification_report(y_test,y_test_hat))
```
